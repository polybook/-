```python
def initial_cond(x, L):
    return math.exp(-(x-L/2)**2/0.02)

L, N = 1, 20
d = L / N

u = np.arange(N+1, dtype=float)
v = np.arange(N+1, dtype=float)
C = 0.5
x = np.linspace(0, L, N + 1)

for i in range(0, N + 1):
  u[i] = initial_cond(x[i], L) *.01

v[0], v[N] = 0, 0
for i in range(0, N -1 ):
  v[i+1] = u[i+1] + C * (u[i+2] - 2 * u[i+1] + u[i])


plt.scatter(x, v)
plt.plot(x, v)
plt.plot(x, u)
```
